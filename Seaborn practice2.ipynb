```python
#let's start the imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from scipy import stats
%matplotlib inline
poke=pd.read_csv('Pokemon 2.csv')
```

```python
#Bivariate

sns.jointplot(x="HP", y="Attack", data=poke);
plt.show()
#HP stands for hidden power, or amount of attack a pokemon can take. Most pokmeon have attacks 50-100.
```

```python
sns.jointplot(x="Total", y="Defense", data=poke, kind="kde")
plt.show()
#Attack and Special Attack. Defense is skewed to th left. 
```

```python
sns.violinplot(poke.Total)
plt.show()
#Total sum of attack shows bimodal distribution
```

```python
ax = sns.scatterplot(x="HP", y="Speed",
                     hue="Generation",
                    sizes=(10, 200),
                     data=poke)
#can't see much here. There is too much to see clearly.
```

```python
sns.set(rc={'figure.figsize':(12,9)})
sns.swarmplot(x = "Type 1", y = "Sp. Def", 
              data = poke)
plt.show()
#gives us a good visual of Type 1 and Special Defense.
```

```python
sns.set(rc={'figure.figsize':(12,9)})
sns.swarmplot(x = "Type 2", y = "Sp. Def", 
              data = poke)
plt.show()
#Type 2 and Special Defense
```

```python
sns.set(rc={'figure.figsize':(12,9)})
sns.swarmplot(x = "Type 1", y = "Attack", 
              data = poke)
plt.title('Type 1 and Attack')
plt.show()
sns.set(rc={'figure.figsize':(12,9)})
sns.swarmplot(x = "Type 2", y = "Attack", 
              data = poke)
plt.title('Type 2 and Attack')
plt.show()
#combined swarmplots. 
```

```python
sns.set(rc={'figure.figsize':(12,9)})
sns.swarmplot(x = "Type 1", y = "Total", 
              data = poke)
plt.title('Type 1 and Total')
plt.show()
sns.set(rc={'figure.figsize':(12,9)})
sns.swarmplot(x = "Type 2", y = "Total", 
              data = poke)
plt.title('Type 2 and Total')
plt.show()
```

```python
g = sns.lmplot(y='Speed', 
               x='Attack',
               hue='Legendary',
               data=poke,
               fit_reg=True,
               ci=False,
               scatter_kws={'alpha':0.4})
plt.show()

g = sns.lmplot(y='Total', 
               x='Attack',
               hue='Legendary',
               data=poke,
               fit_reg=True,
               ci=False,
               scatter_kws={'alpha':0.4})
plt.show()
#Legendary pokemons have higher attacks and speed. Their total sum of attack is also higher. Legendary pokemons are super strong. 
```
