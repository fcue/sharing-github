```python
#let's start the imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
heart=pd.read_csv('heart.csv')
```

```python
heart.head()
```

```python
#Choose one variable and plot that variable four different ways.


f, axes = plt.subplots(2, 2, figsize=(7, 7), sharex=True)
sns.despine(left=True)

d = heart.age

sns.distplot(d, kde=False, color="b", ax=axes[0, 0])

sns.distplot(d, hist=False, rug=True, color="r", ax=axes[0, 1])

sns.distplot(d, hist=False, color="g", kde_kws={"shade": True}, ax=axes[1, 0])

sns.distplot(d, color="m", ax=axes[1, 1])

plt.show()
```

```python
# Choose two continuous variables, and plot them three different ways.
sns.lmplot(x='trestbps', y='chol', data=heart, fit_reg=True, hue='target', height=10)
plt.show()
```

```python
heart.chol.plot(kind="line",color="green",label="chol",grid=True,linestyle=":")
heart.trestbps.plot(kind="line",color="blue",label="trestbps",grid=True)
plt.legend(loc="upper right") #legend: puts feature label into plot
plt.xlabel("indexes")
plt.ylabel("Frequency")
plt.title("Heart Disease Attributes")
plt.show()
```

```python
sns.scatterplot(x="chol", y="trestbps", data=heart, sizes=(10, 200))
plt.show()
```

```python
#Choose one continuous variable and one categorical variable, and plot them six different ways.

sns.catplot(x="target", y="thalach",
            kind="violin",scale='count', inner='quartile', data=heart);
plt.title('Heart Disease and Cholesterol')
plt.show()
#violin plot gives quartiles and show distribution
```

```python
g=sns.catplot(x="target", y="thalach", kind="swarm", legend_out=True, data=heart)
plt.show()
#nice visual with a swarmplot
```

```python
sns.boxplot(x="target", y="thalach",  data=heart)
plt.show()
#similar to violin plot, shows n
```

```python
sns.stripplot(x="target", y="thalach",
              data=heart, dodge=True, jitter=False,
              alpha=0.8, zorder=1)
plt.show()
#not good visual for these variables
```

```python

sns.set(style="ticks")
sns.lmplot(x="target", y="thalach", data=heart,
            ci=None, palette="muted", height=8,
           scatter_kws={"s": 80, "alpha": 1})
plt.show()
```

```python
sns.barplot(x=heart.target, y=heart.thalach, palette="pastel")
plt.show()
#would be better if there was visible numbers on bars because values are close
```
